```python
song_count_by_artist = data
song_count_by_artist["trackCountByArtist"] = data.groupby(["artistName"])["msPlayed"].transform('count')
song_count_by_artist = song_count_by_artist.drop_duplicates(subset=["artistName"])
song_count_by_artist = song_count_by_artist.sort_values(by="trackCountByArtist", ascending=False)
song_count_by_artist = song_count_by_artist.drop(["endTime", "trackName","msPlayed"], axis=1)


%store song_count_by_artist
# You can find how many times I listened songs based on Artist = new_data 
# Does not mean that songs are not repeated

# so if we take "The Cure" whos trackCountByArtist = 273, it means that I turned on 273 cure songs
# (maybe it was 5 songs but 273 means that music started 273 times)
```

```python
top_50_songs = data
top_50_songs["trackCount"] = top_50_songs.groupby(["trackName"])["msPlayed"].transform('count')
top_50_songs = top_50_songs.drop_duplicates(subset=["trackName"])
top_50_songs = top_50_songs.drop(["endTime", "msPlayed", "trackCountByArtist"], axis=1)
top_50_songs = top_50_songs.sort_values(by="trackCount", ascending=False).iloc[:50]

%store top_50_songs
# You can find the sorted trackCount by trackName, meaning how many times I listened for one song = top_50_songs 
```
